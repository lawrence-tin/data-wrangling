# Data Wrangling

#### before we can answer any business question, we need to clean the data to make sure it makes sense

#### Import Libraries

```python
import pandas as pd
import matplotlib.pylab as plt
```

#### Import data

```python
df = pd.read_csv('auto.csv')
df.head() # let's view the structure of the data by showing the first 5 rows
```

As can be seen from the dataset head, there are no headers, so we don't know what each column mean, so let's define headers

#### define headers for the data 

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

Let's add column names to the dataset as we defined above

```python
df.columns = headers
```

```python
# To see what the data set looks like, we'll use the head() method.
df.head()
```

## Handling Missing values

We also noticed that there are some missing values from the dataset, and we need to solve that before we begin any analysis

```python
import numpy as np

# replace "?" to NaN, a language understood by Python
df.replace("?", np.nan, inplace = True)
df.head(5)
```

### Evaluating for Missing Data

```python
# Let's see how many missing values are available in each column of the dataframe
df.isnull().sum()
```

#### Let's see how big is our dataset so we can decide on what to do with missing values

```python
df.shape
```

It seems it is safe to replace missing values with a column mean

#### Calculate the average of the column

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0) #average of normalized-losses column
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True) #replace missing values with mean
```

```python
avg_bore = df['bore'].astype('float').mean(axis=0) #average of bore column
df["bore"].replace(np.nan, avg_bore, inplace=True) #replace missing values with mean
```

```python
avg_stroke = df['stroke'].astype('float').mean(axis=0) #average of stroke column
df["stroke"].replace(np.nan, avg_stroke, inplace=True) #replace missing values with mean
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0) #average of horsepower column
df["horsepower"].replace(np.nan, avg_horsepower, inplace=True) #replace missing values with mean
```

```python
avg_peak_rpm = df['peak-rpm'].astype('float').mean(axis=0) #average of peak-rpm column
df["peak-rpm"].replace(np.nan, avg_peak_rpm, inplace=True) #replace missing values with mean
```

```python
avg_price = df['price'].astype('float').mean(axis=0) #average of peak-rpm column
df["price"].replace(np.nan, avg_price, inplace=True) #replace missing values with mean
```

For 'num-of-doors' column, let's replace missing values with mode since it's non-value column

```python
df['num-of-doors'].value_counts()
```

```python
#replace the missing 'num-of-doors' values by the most frequent 
df["num-of-doors"].replace(np.nan, "four", inplace=True) # Four is the most frequent (114)
```

#### Let's see if our dataframe has no missing values

```python
df.isnull().sum()
```

#### for analysis, we have to make sure all data in every column is in the right type

```python
df.dtypes
```

#### We discovered that some column values can't be processed in the form they are in, so we need to transform the data

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int") # losses are supposed to be integers
df[["price"]] = df[["price"]].astype("float") # price is always a float
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float") # rotation is always a float
print('Data Transformed')
```

Let's find out the data types after transformation

```python
df.dtypes
```

```python
df.head()
```

### Data Normalization

#### To avoid overprocessing of some columns, let's transform data values to similar range

```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
 df['height'] = df['height']/df['height'].max()
```

```python
df.head()
```

## Indicator variable (or dummy variable)

#### for continuous analysis, we need to convert all values to numbers, so we get dummy variables to replace each non-numerical value

```python
df.columns
```

#### For 'fuel-type'

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

#### Replace the 'fuel-type column with newly generated columns, to reduce confusion

```python
dummy_variable_1.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```

#### For 'aspiration'

```python
dummy_variable_2 = pd.get_dummies(df["aspiration"])
dummy_variable_2.head()
```

#### Replace the 'aspiration' column with newly generated columns, to reduce confusion

```python
dummy_variable_2.rename(columns={'aspiration':'std', 'aspiration':'turbo'}, inplace=True)
dummy_variable_2.head()
```

```python
# merge data frame "df" and "dummy_variable_2" 
df = pd.concat([df, dummy_variable_2], axis=1)

# drop original column "aspiration" from "df"
df.drop("aspiration", axis = 1, inplace=True)
```
